```python
import pandas as pd
```

```python
dataInFolder = './../data/'
rentInfo = pd.read_csv(dataInFolder + '공공 자전거 대여소 정보(22.06월 기준).csv', encoding = 'UTF-8')
rentInfo.head()
```

```python
rentInfo.columns
```

```python
rentInfo['대여소이름'] = rentInfo['대여소번호'].astype('str') + '.' + rentInfo['대여소명']
rentInfo = rentInfo.drop(['대여소번호', '대여소명'], axis=1)
rentInfo = rentInfo.set_index('대여소이름')
rentInfo.head()
```

```python
rentInfo.dtypes.to_frame().T
```

```python
result = rentInfo.groupby('자치구')['자치구']
result = result.count().sort_values(ascending=False).head()
result.to_frame()
```

<h2><font color="red">서울시 공공 자전거 실시간 대여 정보(open API로 json 데이터 가져옴)</font></h2>

```python
dataInFolder = './../data/'
bikeList = pd.read_csv(dataInFolder + 'bikeList.csv', encoding = 'UTF-8', index_col=0)

bike_dict = {'rackTotCnt':'거치대개수', 'stationName':'대여소이름', \
             'parkingBikeTotCnt':'자전거주차총건수', 'shared':'거치율', \
             'stationLatitude':'위도', 'stationLongitude':'경도', 'stationId':'대여소ID'}

bikeList = bikeList.rename(columns=bike_dict)
bikeList.head()
```

```python
bikeList['대여소이름'] = bikeList['대여소이름'].str.replace('. ', '.', regex=False) 
bikeList = bikeList.set_index('대여소이름')
bikeList.head()
```

```python
bikeList.columns
```

```python
# '거치율'이 가장 높은 항목 5개
result = bikeList['거치율'].sort_values(ascending=False).head()
result.to_frame()
```

```python
result.plot(kind='barh', grid=False, use_index=True, rot=0, \
                color=['r', 'b', 'g', 'DarkMagenta', 'Gold'])
plt.xlabel('거치율')
plt.ylabel('대여소 이름')
plt.title('거치율 상위 top 5')
```

```python
# 두 데이터 프레임 합치기
mergedata = pd.merge(rentInfo, bikeList, left_index=True, right_index=True, \
                    how='outer', indicator=True, suffixes=('', '_right'))
mergedata = mergedata.drop(['위도_right', '경도_right'], axis=1)
mergedata.index.name = '대여소이름'
mergedata.columns
```

```python
# '자치구'별 '거치대 개수'의 합 구하고, 상위 1개만 그래프 그리기
result = mergedata.groupby('자치구')['거치대개수']
result = result.sum().sort_values(ascending=False).head(10)
result.to_frame()
```

```python
result.plot(kind='barh', grid=False, use_index=True, rot=0, \
                color=['r', 'b', 'g', 'DarkMagenta', 'Gold'])
plt.xlabel('거치대 개수')
plt.ylabel('대여소 이름')
plt.title('거치대 개수 상위 top 10')
```

<h2><font color="red">서울시 공공 자전거 이용 정보(일별)</font></h2>

```python
import pandas as pd
```

```python
dataInFolder = './../data/'
cycleDayRent = pd.read_csv(dataInFolder + 'cycleRentUseDayInfo(20221001~20221010).csv' , 
                       index_col=0,encoding='UTF-8')

day_rent_dict = {'RENT_DT':'대여일자(YYYYMMDD)','RENT_ID':'대여소번호',
                 'RENT_NM':'대여소명','RENT_TYPE':'대여구분코드','GENDER_CD':'성별',
                 'AGE_TYPE':'연령','USE_CNT':'대여건수','EXER_AMT':'운동량','CARBON_AMT':'탄소절감량',
                 'MOVE_METER':'이동거리(M)','MOVE_TIME':'이용시간(분)'
}

cycleDayRent = cycleDayRent.rename(columns = day_rent_dict)

#전처리
cycleDayRent['대여소명'] = cycleDayRent['대여소명'].str.replace('. ' , '.' , regex=False)
cycleDayRent['연령'] = cycleDayRent['연령'].str.replace('~' , '' , regex=False)
cycleDayRent['성별'] = cycleDayRent['성별'].str.upper()

# 대여소명을 색인으로 보내기
cycleDayRent = cycleDayRent.set_index('대여소명')
cycleDayRent.head()
```

```python
# '대여구분코드' 별로 그룹핑 후 파이 그래프 만들기
print(cycleDayRent['대여구분코드'].unique())
chartdata = cycleDayRent['대여구분코드'].value_counts()
chartdata.to_frame()
```

```python
import matplotlib.pyplot as plt

ratio = [411826
, 129995, 6808, 5797]
labels = ['정기권', '일일권', '일일권(비회원)', '단체권']
explode = [0.05, 0.05, 0.05, 0.05]
colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0']

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode, shadow=True, colors=colors)
plt.show()
```

```python
# '연령' 별로 막대 그래프 만들기
print(cycleDayRent['연령'].unique())
chartdata = cycleDayRent['연령'].value_counts().sort_index()
chartdata.to_frame()
```

```python
# 연령별 빈도수 막대 그래프로 그리기
chartdata.plot(kind='bar', rot=0,
              use_index=True, table=False, color=['r','g','b','y','m'])

plt.xlabel['빈도수']
plt.ylabel['연령대']
plt.title['연령별 빈도 수']
```

```python
# '성별','연령','대여구분코드','이동거리(M)'를 이용한 pivoting
# '연령' 에서 '기타' 항목을 제외 시킵니다.
# '성별'이 NaN인 행들은 항목에서 제외 시킵니다.
concern= ['성별','연령','대여구분코드','이동거리(M)']
concernData = cycleDayRent[concern]

print('성별 before delete : ' + str(len(concernData)))
concernData = concernData.dropna(subset=['성별'])
print('성별 after delete : ' + str(len(concernData)))

concernData = concernData[concernData['연령'] != '기타']
print('\'연령\' != \'기타\' after delete : ' + str(len(concernData)))

concernData.head()
```

```python
pivotData = concernData.pivot_table(index=['성별','연령'], 
                                    columns='대여구분코드', 
                                    values='이동거리(M)', 
                                    aggfunc='mean')

pivotData = pivotData.fillna(value={'일일권(비회원)':0}, axis=0, inplace=False)
pivotData
```

```python
# MultiIndex를 SingleIndex로 변경하기
outer_index = pivotData.index.get_level_values(0)
outer_index = pd.Index(['여자' if idx=='F' else '남자' for idx in outer_index]) #lsit complination


inner_index = pivotData.index.get_level_values(1)
outer_index, inner_index

newindex = outer_index + '_' + inner_index
pivotData.index = newindex
pivotData
```

```python
# 일일권 과 일일권(비회원)을 합쳐서 일일권에 묶습니다.
pivotData['일일권'] = pivotData['일일권'] + pivotData['일일권(비회원)']
pivotData = pivotData.drop('일일권(비회원)', axis = 1)
pivotData
```

```python
#색인 14개가 pivotData.index이 된다.
womanidx = [item for item in pivotData.index if item.startswith('여자')] 
woman = pivotData.loc[womanidx]
woman #여자들 정보
```

```python
manidx = [item for item in pivotData.index if item.startswith('남자')] 
man = pivotData.loc[manidx]
man #남자들 정보
```

```python
# 여성별 연령별 대여구분코드별 평균 막대 그래프

woman.plot(kind='bar', rot=0, 
           title='여성별 연령별 대여구분코드별 평균 막대 그래프',
           legend=True)

plt.legend(loc='best') # 범례는 적절한 위치에 두어라(best)
plt.xlabel('연령대')
plt.ylabel('평균')
```

```python
# woman.T.plot (행과 열을 전치하여 그리기))

woman.T.plot(kind='bar', rot=0, 
           title='대여구분코드별 연령별 평균 막대 그래프(여성)',
           legend=True)

plt.legend(loc='best') 
plt.xlabel('대여구분코드')
plt.ylabel('평균')
```

```python
# 데이터 프레인 병합하기
# 두 데이터 프레임 합치기
mergedata02= pd.merge(rentInfo, cycleDayRent, left_index=True, right_index=True, \
                    how='outer', indicator=True, suffixes=('', '_right'))

mergedata02.columns
```

```python
# '자치구'별 '대여구분코드'별로 '대여건수'의 총합 구하기
concern = ['자치구','대여구분코드','대여건수']
bikeData = mergedata02[concern]
bikeData.head()
```

```python
groupedData = bikeData.groupby(['자치구','대여구분코드'])['대여건수']
result = groupedData.sum().to_frame().reset_index()
result = result.sort_values(by=['자치구','대여구분코드'],ascending=[True, False])
result.head(20)
```

```python
# 각 '자치구'별로 '대여구분코드'의 최대 값을 구해 봅니다.
result = result.drop_duplicates(subset=['자치구'])
result.head()
```
