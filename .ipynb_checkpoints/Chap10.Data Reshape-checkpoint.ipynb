```python
import pandas as pd
```

```python
myframe01 = pd.read_csv('./../data/payment01.csv', encoding='cp949')
myframe01
```

```python
myframe02 = pd.read_csv('./../data/payment02.csv', encoding='cp949')
myframe02
```

```python
columninfo01 = myframe01.columns
columninfo02 = myframe02.columns
columninfo01.intersection(columninfo02)
```

```python
columninfo01.difference(columninfo02)
```

```python
columninfo02.difference(columninfo01)
```

```python
concatRow = pd.concat([myframe01,myframe02], axis=0)
concatRow
```

```python
concatRow.iloc[3].to_frame()
```

```python
newFrame = pd.DataFrame([['이순철','인천',3.0,'영업부','남자',1000.0,
                         5000,2000.0,100.0]],
                       columns=['사원명','출장지역','출장기간','부서',
                                '성별','교통비','식비','숙박비','부대비용'])
newFrame
```

```python
result = pd.concat([myframe01, newFrame])
result
```

```python
concatRow = pd.concat([myframe01, myframe02], axis=0, ignore_index=True)
concatRow
```

```python
col_concat = pd.concat([myframe01, myframe02], axis=1)
col_concat
```

```python
myframe01.columns = ['사원명', '출장지역', '출장기간', '부서01', '성별01', 
                     '교통비01', '식비01']
```

```python
myframe02.columns = ['사원명', '출장지역', '출장기간', '부서02', '성별02', 
                     '교통비02', '식비02', '숙박비02', '부대비용02']
```

```python
col_concat = pd.concat([myframe01, myframe02], axis=1)
col_concat
```

```python
result = pd.concat([myframe01, myframe02], join='inner', ignore_index=True)
result
```
