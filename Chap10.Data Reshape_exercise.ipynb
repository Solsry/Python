<h3>merge</h3>

<h4>중제목</h4>

<h6>소제목</h6>

```python
import pandas as pd
import numpy as np
```

<h4><font color="green">문제01</font></h4>

```python
file01 = './../data/member01.csv'
member = pd.read_csv(file01, encoding='cp949')
member
```

```python
file11 = './../data/board11.csv'
board11 = pd.read_csv(file11, encoding='cp949')

file12 = './../data/board12.csv'
board12 = pd.read_csv(file12, encoding='cp949')

board = pd.concat([board11, board12])
board
```

```python
# 1. 양쪽에 공통으로 들어 있는 데이터를 조회해 보세요.
result = pd.merge(member, board, on='아이디')
result
```

```python
# 2. 양쪽에 공통으로 들어 있는 데이터를 조회하되, 왼쪽에만 들어 있는 데이터도 같이 조회해 보세요.
result = pd.merge(member, board, on='아이디', how='left')
result
```

```python
# 3. 양쪽에 공통으로 들어 있는 데이터를 조회하되, 오른쪽에만 들어 있는 데이터도 같이 조회해 보세요.
result = pd.merge(member, board, on='아이디', how='right')
result
```

```python
# 4. 양쪽에 공통으로 들어 있는 데이터와 한쪽에만 들어 있는 데이터도 같이 조회해 보세요.
result = pd.merge(member, board, on='아이디', how='outer')
result
```

<h4><font color="green">문제03</font></h4>

```python
file01 = './../data/member01.csv'
member = pd.read_csv(file01, encoding='cp949', index_col='아이디')
member
```

```python
file11 = './../data/board11.csv'
board11 = pd.read_csv(file11, encoding='cp949')

file12 = './../data/board12.csv'
board12 = pd.read_csv(file12, encoding='cp949')

board = pd.concat([board11, board12])
board = board.set_index('아이디')
board
```

```python
# 1. 양쪽에 공통으로 들어 있는 데이터를 조회해 보세요.
result = pd.merge(member, board, left_index=True, right_index=True)
result
```

```python
# 2. 양쪽에 공통으로 들어 있는 데이터를 조회하되, 왼쪽에만 들어 있는 데이터도 같이 조회해 보세요.
result = pd.merge(member, board, left_index=True, right_index=True, how='left')
result
```

```python
# 3. 양쪽에 공통으로 들어 있는 데이터를 조회하되, 오른쪽에만 들어 있는 데이터도 같이 조회해 보세요.
result = pd.merge(member, board, left_index=True, right_index=True, how='right')
result
```

```python
# 4. 양쪽에 공통으로 들어 있는 데이터와 한쪽에만 들어 있는 데이터도 같이 조회해 보세요.
result = pd.merge(member, board, left_index=True, right_index=True, how='outer')
result
```

<h3>형상 변경하기</h3>

<h4>중제목</h4>

```python
import pandas as pd
```

```python
filename = './../data/student_score.csv'
score = pd.read_csv(filename, encoding='CP949')
score.head()
```

```python
score_pivot = score.pivot(index=['name', 'gender'], columns='subject', values='jumsu')
score_pivot
```

<h6>melt() 메소드 사용</h6>

```python
sungjuk = pd.read_csv('./../data/sungjuk_sheet.csv', encoding='CP949')
sungjuk = sungjuk.set_index(['name', 'gender'])
sungjuk
```

```python
data = pd.concat([score_pivot, sungjuk])
data = data.reset_index()
data
```

```python
data_long = pd.melt(data, id_vars =['name', 'gender'],
                var_name='과목', value_name = '점수')
data_long.head()
```

```python
data_tidy = data_long.pivot_table(
    index=['gender'], columns=['과목'], values='점수', aggfunc=['sum', 'max'] ) 
data_tidy
```

```python
data_tidy = data_long.pivot_table(
    index=['과목'], columns=['gender'], values='점수', aggfunc=['min', 'mean'] ) 
data_tidy
```
