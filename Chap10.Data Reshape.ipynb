```python
import pandas as pd
import numpy as np
```

```python
file01 = './../data/학생정보01.csv'
student = pd.read_csv(file01, encoding='cp949')
student
```

```python
file02 = './../data/시험점수01.csv'
jumsu = pd.read_csv(file02, encoding='cp949')
jumsu
```

```python
col_student = student.columns
col_jumsu=jumsu.columns
col_student.intersection(col_jumsu)
```

```python
result = pd.merge(student, jumsu, on='id') 
result
```

```python
result = pd.merge(student, jumsu, on='id', how='outer')
result
```

```python
result = pd.merge(student, jumsu, on='id', how='left')
result
```

```python
result = pd.merge(student, jumsu, on='id', how='right')
result
```

<h6><font color="BlueViolet">중첩 색인된 데이터 다루기

</font></h6>

```python
left_frame = pd.DataFrame({'지역':['서울','서울','서울','부산','부산'],
                          '년도':[2018,2019,2020,2019,2020],
                          'data':[10 * idx for idx in range(1,6)]})
left_frame
```

```python
myindex = [
    ['부산', '부산', '서울', '서울', '서울', '서울'], 
    [2022, 2021, 2021, 2021, 2022, 2023]
]
mycolumn = ['data01', 'data02']

data = np.arange(12).reshape((6, 2))

right_frame = pd.DataFrame(data, index=myindex, columns=mycolumn)
right_frame
```

```python
result = pd.merge(left_frame, right_frame, left_on=['지역','년도'],
                 right_index=True, how='outer')
result
```

```python
movshow = pd.read_csv('./../data/movie_show.csv', encoding='cp949')
movshow
```

```python
movshow_melt = pd.melt(movshow, id_vars=['gu','year','month'], var_name='genre',value_name='qty')
movshow_melt
```

```python
movshow_tidy = movshow_melt.pivot_table(index=['gu','year'],
                                       columns='month',values='qty')
movshow_tidy
```

```python
movshow_tidy = movshow_melt.pivot_table(index=['gu','year'],
                                       columns='month', values=['genre','qty'], 
                                        aggfunc={'genre':'count' , 'qty':'mean'})
movshow_tidy
```

<h6><font color="BlueViolet">3)항목별 다른 함수 적용하기
</font></h6>

```python
import numpy as np
```

```python
movshow_tidy = movshow_melt.pivot_table(index=['gu','year'],
                                       columns='genre',values='qty')
movshow_tidy
```

```python
movshow_tidy = movshow_melt.pivot_table(index=['gu','year'],
                                       columns='genre',values='qty', aggfunc='sum')
movshow_tidy
```

```python
movshow_tidy = movshow_melt.pivot_table(index=['gu','year'],
                                       columns='genre',values='qty', aggfunc={'qty':['mean','sum']})
movshow_tidy
```

```python
outer = movshow_tidy.columns.get_level_values(0) # 외부 색인
inner = movshow_tidy.columns.get_level_values(1) # 내부 색인
newindex = inner + '_' + outer
movshow_tidy.columns = newindex
movshow_tidy
```

```python
result = movshow_tidy.reset_index()
result
```
