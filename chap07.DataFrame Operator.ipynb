<h6><font color="BlueViolet">2) 메모리 사용량 확인
</font></h6>

```python
print(payment['출장기간'].max())
print(payment['출장기간'].min())
```

```python
before_memory = payment.memory_usage(deep=True)
before = before_memory.to_frame().T
before
```

```python
payment['출장기간'] = payment['출장기간'].astype(np.int8)
after_memory = payment.memory_usage(deep=True)
after = after_memory.to_frame().T
after
```

```python
newindex = ['before','after']
result = pd.concat([before,after])
result.index = newindex
result
```

```python
ratio = (after_memory / before_memory).to_frame().T
ratio
```

<h3><font color="green">상위 x개에서 하위 y개 고르기

<h6><font color="BlueViolet">1) 상위 x명 추출

```python
top = 4
print('부대비용 상위 ' + str(top) + '명 추출하기')
payment.nlargest(top, '부대비용')
```

<h6><font color="BlueViolet">2) 상위 x명 에서 하위 y명 추출
</font></h6>

```python
bottom = 2
payment.nlargest(top, '부대비용').nsmallest(bottom , '숙박비')
```
