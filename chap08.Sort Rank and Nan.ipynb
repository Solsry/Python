```python
import numpy as np
import pandas as pd
```

```python
payment = pd.read_csv('./../data/payment01.csv', encoding ='cp949',
                     index_col='사원명')
payment
```

<h6><font color="BlueViolet">1) 스칼라 값을 이용한 누락된 값 치환

</font></h6>

```python
# inplace=True 면 오리지널 원본 까지 바꿔주고, False면 NaN 그대로 살아 있음.
result = payment.fillna(value=0, inplace=False)
result
```

```python
payment
```

```python
payment.fillna(value=0, inplace=True)
payment
```

```python
payment
```

```python
payment.loc[['강감찬','김유신'],['출장기간','성별']] = np.nan
payment.loc[['신사임당','선덕여왕'],['식비']] = np.nan
payment
```

<h6><font color="BlueViolet">2) 사전을 이용한 누락 값 치환

</font></h6>

```python
mydict = {'출장기간':10,'성별':'남자','식비':500.0}
payment.fillna(mydict , inplace=True)
payment
```

<h6><font color="BlueViolet">3) 누락된 값 치환

</font></h6>

```python
payment.loc[['신사임당'],['출장기간']] = np.nan
payment.loc[['신사임당'],['교통비']] = np.nan
payment.loc[['선덕여왕'],['식비']] = np.nan
payment
```

```python
mydict = {'출장기간':payment['출장기간'].mean(),
         '교통비':payment['교통비'].mean(),
         '식비':payment['식비'].mean(),}

payment.fillna(mydict, inplace=True)
payment
```
